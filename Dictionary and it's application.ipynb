## Dictionary

1. A dictionary is one of the collection data type python offers.
2. Dictionary exist in key value pairs.
3. { key: value }

Dictionary differ from list how elements are accessed.
1. List elements are accessed by index numbers.
2. Dictionary elements are accessed by their keys.

1. You can define a dictionary by enclosing a comma-separated list of key-value pairs in curly braces ({})
2. A colon (:) separates each key from its associated value

```python
# How we difine a dictionary
dict1 = { 'name' : 'Sachin',
          'name1': 'Rahul',
          'name2': 'Nikita'
        }
```

Difference between list and dictionary.


```python
# Define a dictionary using dictionary constructor.
dict2 = dict([
             ('name' , 'Sachin'),
             ('name1' ,  'Rahul'),
             ('name2' , 'Nikita')
            ])
```

```python
print(dict1)
print(dict2)
# the output of both dict1 and dict2 will be same. 
```
